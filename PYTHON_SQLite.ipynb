**SQL**

A Linguagem SQL (Structured Query Language) é uma linguagem de programação padrão usada para gerenciar e manipular bancos de dados relacionais. Ela permite aos usuários acessar, modificar, inserir e excluir dados, bem como criar e gerenciar tabelas e estruturas de banco de dados.

O SQL  é  amplamente  utilizado  por  Desenvolvedores,  Analistas  de  Dados,  Cientistas  de Dados, Engenheiros de Dados e Administradores de Banco de Dados (DBAs) para trabalhar com bancos de dados relacionais, como MySQL, PostgreSQL, Oracle, SQL Server, DB2, SQLite, entre outros.

A Linguagem SQL consiste em várias instruções, que podem ser agrupadas nas seguintes categorias:

1-DDL (Data Definition Language) - define e gerencia a estrutura de tabelas e objetos do banco de dados:

* CREATE: cria uma novatabela ou objeto no banco de dados.

* ALTER: modifica uma tabela ou objeto existente.

* DROP: exclui uma tabela ou objeto existente.

2-DML (Data Manipulation Language) - manipula os dados armazenados nas tabelas:
* SELECT: recupera dados de uma ou mais tabelas.

* INSERT: insere novos registros em uma tabela.

* UPDATE: atualiza registros existentes em uma tabela.

* DELETE: exclui registros de uma tabela.

3-DCL (Data Control Language) - gerencia permissões e controle de acesso aos objetos do banco de dados:

* GRANT: concede permissões a usuários ou grupos para acessar e manipular objetos do banco de dados.

* REVOKE: remove permissões concedidas anteriormente.

4-TCL (Transaction Control Language) - gerencia transações no banco de dados:

* BEGIN TRANSACTION: inicia uma nova transação.

* COMMIT: salva todas as alterações feitas na transação atual.

* ROLLBACK: desfaz as alterações feitas na transação atual.

* SAVEPOINT:  cria  um  ponto  de  salvamento  na  transação  atual,  permitindo  que  você reverta para esse ponto, se necessário.

**SGBD**

Um  Sistema  Gerenciador  de  Banco  de  Dados  (SGBD)  é  um  software  projetado  para gerenciar,  armazenar,  manipular  e  consultar  dados  em  um  banco  de  dados.  SGBDs  são fundamentais  para  o  gerenciamento  eficiente  e  seguro  de  grandes  quantidades  de  dados  em aplicações e sistemas de informação.Os SGBDs fornecem uma interface entre os usuários e os dados armazenados, permitindo a realização de diversas operações, como:

* Criação e modificação de estruturas de dados (tabelas, índices, etc...).
* Inserção, atualização e exclusão de dados.
* Realização de consultas e recuperação de informações.
* Controle de acesso e segurança, garantindo que apenas usuários autorizados possam acessar e modificar os dados.
* Garantia de integridade e consistência dos dados.
* Otimização de consultas e melhoria do desempenho do banco de dados.
* Recuperação de falhas e realização de backups.

Existem  diversos  tipos  de  SGBDs  disponíveis  no  mercado,  como:
* Os  Relacionais  (MySQL, PostgreSQL,  Microsoft  SQL  Server,  Oracle,  IBM  DB2,  SQLite,  etc...)
* Os Orientados  a  Objetos (MongoDB, Couchbase, etc...)
* Os de Colunas (Apache Cassandra, HBase, etc...)
* entre outros...

A escolha  do  SGBD  adequado  depende  das  necessidades  e  requisitos  específicos  do  projeto  ou aplicação.

Um  SGBD  pode  gerenciar  um  ou  mais  bancos  de  dados.  Os  bancos  de  dados  são  as estruturas onde os dados são realmente armazenados.

## Conectando no Banco de Dados com Linguagem Python

```python
# Conecta no banco de dados
con = sqlite3.connect('cap12_dsa.db') # Nome do arquivo
# con = objeto do tipo 'connect'
# Obs: o arquivo deve estar na mesma pasta que o notebook
```

```python
# Abre um cursor para percorrer os dados no banco de dados
cursor = con.cursor()

# Cursor: espécie de canal para vc manipular os dados daquele banco
```

```python
# Query SQL para extrair os nomes das colunas no banco de dados
sql_query = """SELECT name FROM sqlite_master WHERE type = 'table';"""

# query = consulta
# letras em maiúsculo - sintaxe SQL
# SELECT - selecionar, na coluna 'name'
# FROM - a partir de um lugar... tabela 'sqlite_master'
# WHERE - na qual... (filtro)... onde o 'type' = 'table'... ou seja, quais são as tabelas desse BD
# Obs: o 'table' está com aspas simples... logo, a query tem que estar entre aspas duplas 3x de cada lado (doc string)

# finaliza a query com ';' no final
```

```python
# Executa a query
cursor.execute(sql_query)
# execute = executar a query
# o resultado está na memória do computador
```

```python
# Visualiza o resultado
print(cursor.fetchall())

# fetchall - pegar o resultado
```

> A query abaixo retorna a média de unidades vendidas por produto.

```python
# Cria uma instrução SQL para calcular a média de unidades vendidas por produto
# Média por produto vendido
query3 = 'SELECT Nome_Produto, AVG(Unidades_Vendidas) FROM tb_vendas_dsa GROUP BY Nome_Produto'
# AVG = média...
# Faça o SELECT do 'Nome_Produto', calcule a média 'Unidades_Vendidas',
# a partir da tabela 'tb_vendas_dsa'
# agrupando (GROUP BY) por 'Nome_Produto'

# OBs: a coluna que não está na função de agregação, vai para o Group By
```

```python
# Visualiza o resultado
cursor.fetchall()
```

OBS: ESSA SINTAXE SQL É A 'SQL ANSI', FORMATO GLOBAL, PADRÃO.

> A query abaixo retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199.

```python
# Cria uma instrução SQL para calcular a média de unidades vendidas por produto,
# quando o valor unitário for maior que 199
query4 = """SELECT Nome_Produto, AVG(Unidades_Vendidas)
            FROM tb_vendas_dsa
            WHERE Valor_Unitario > 199
            GROUP BY Nome_Produto"""

# para quebrar em linhas a instrução, usar os 3 """ ... """
```

```python
# Executa a query no banco de dados
cursor.execute(query4)
```

```python
# Visualiza o resultado
cursor.fetchall()
```

> A query abaixo retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199 e somente se a média de unidades vendidas for maior do que 10.

```python
# Esta query está ERRADA!

query5 = """SELECT Nome_Produto, AVG(Unidades_Vendidas)
            FROM tb_vendas_dsa
            WHERE Valor_Unitario > 199 and AVG(Unidades_Vendidas) > 10
            GROUP BY Nome_Produto """

            # aplicando 2 filtros
```

```python
# Executa a query no banco de dados
cursor.execute(query5)

# vai dar erro no segundo 'AVG'...
# a query de SQL tem uma ordem de execução das cláusulas SQL!
```

**Ordem**
* 1º: **FROM**... buscar os dados na tabela
* 2º: aplica o **WHERE** (filtro)
* 3º: depois pegar as colunas (**SELECT**)
* 4º: fazer o agrupamento (**GROUPBY**)

Obs: o AVG só é calculado depois do Groupby.

O erro acima é devido a tentativa de usar o 'avg' antes do 'groupby'

**NÃO PODE USAR O 'AVG' NA CLÁUSULA 'WHERE'**. É uma regra!

```python
# Cria uma instrução SQL para calcular a média de unidades vendidas por produto,
# quando o valor unitário for maior que 199, mas somente se a média de unidades vendidas for maior que 10
query5 = """SELECT Nome_Produto, AVG(Unidades_Vendidas)
            FROM tb_vendas_dsa
            WHERE Valor_Unitario > 199
            GROUP BY Nome_Produto
            HAVING AVG(Unidades_Vendidas) > 10"""

            # HAVING - Cláusula para usar o filtro depois do 'GroupBy'
```

Ou seja...
* WHERE...
* GROUPBY...
* HAVING...

```python
# Executa a query no banco de dados
cursor.execute(query5)
```

```python
# Visualiza o resultado
cursor.fetchall()
```

> A query abaixo retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199.

```python
# Retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199.
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto')['Unidades_Vendidas'].mean()
# '[]'- slicing, fatiar o dataframe
# df['Valor_Unitario'] > 199 -- filtro
# depois, groupby
```

> A query abaixo retorna a média de unidades vendidas por produto se o valor unitario for maior do que 199 e somente se a média de unidades vendidas for maior do que 10.

```python
# Alternativa A
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto').filter(lambda x: x['Unidades_Vendidas'].mean() > 10)
```

```python
# Alternativa B
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto') \
                              .filter(lambda x: x['Unidades_Vendidas'].mean() > 10) \
                              .groupby('Nome_Produto')['Unidades_Vendidas'].mean()


# '\' contra-barra : para cortar o código
# Diz para o interpretador que está continuando na linha seguinte
```

## Sintaxe SQL x Sintaxe Pandas

As duas instruções abaixo retornam o mesmo resultado!

```python
# Sintaxe SQL
query5 = """SELECT Nome_Produto, AVG(Unidades_Vendidas)
            FROM tb_vendas_dsa
            WHERE Valor_Unitario > 199
            GROUP BY Nome_Produto
            HAVING AVG(Unidades_Vendidas) > 10"""

# Pode usar em qualquer SGBD!
# Mais simples de usar! (Sintaxe)
```

```python
# Sintaxe Pandas
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto') \
                              .filter(lambda x: x['Unidades_Vendidas'].mean() > 10) \
                              .groupby('Nome_Produto')['Unidades_Vendidas'].mean()

# Usar apenas no Python.
# Código... mais flexível do SQL!
# Sintaxe é mais rápida... se o BD for grande vai demorar... O Pandas são otimizadas
```

## Links

Ordem de Execução Detalhada:
1. **FROM/JOIN:**
Especifica as tabelas envolvidas na consulta e como elas são unidas (se houver junções).
2. **WHERE:**
Filtra as linhas da tabela ou tabelas especificadas, aplicando condições para selecionar apenas as linhas relevantes.
3. **GROUP BY:**
Agrupa as linhas com base nos valores de uma ou mais colunas, criando grupos de dados.
4. **HAVING:**
Filtra os grupos criados pelo GROUP BY, aplicando condições aos resultados agrupados.
5. **SELECT:**
Seleciona as colunas que serão incluídas no resultado final da consulta.
6. **DISTINCT:**
Remove linhas duplicadas do resultado, garantindo que cada linha seja única.
7. **ORDER BY:**
Ordena o resultado da consulta com base em uma ou mais colunas, em ordem crescente ou decrescente.
8. **LIMIT/OFFSET:**
Restringe o número de linhas retornadas pela consulta e/ou define um deslocamento para começar a retornar resultados a partir de uma determinada posição.

Entender a ordem de execução ajuda a:
* Escrever consultas mais eficientes: Otimizando a ordem das cláusulas, é possível reduzir o processamento desnecessário.
* Depurar consultas: Identificar problemas de desempenho ou erros na lógica da consulta.
* Entender como os resultados são obtidos: A ordem de execução explica o porquê de certos resultados serem retornados e outros não.
