# Deep Learning

"**Deep  Learning**  é  um  subcampo  da  Inteligência  Artificial  (IA)  e  aprendizado  de  máquina que se concentra no desenvolvimento de algoritmos e modelos que imitam a maneira como o cérebro humano processa informações e aprende a partir de dados. Esses modelos são chamados de  redes  neurais  artificiais,  que  consistem  em  várias  camadas  de  neurônios  matemáticos interconectados para processar informações de maneira hierárquica e não linear.

O  termo  "Deep" em  Deep  Learning refere-se  à  presença  de  várias  camadas  ocultas  em umarede neural, o que permite ao modelo aprender representações mais complexas e abstratas dos dados. Isso contrasta com o aprendizado raso (shallow learning), que envolve modelos com apenas  uma  ou  poucas  camadas  ocultas  e,  portanto,  é  limitado  na  capacidadede  aprender representações complexas.

Deep Learning tem sido aplicado com sucesso em uma ampla gama de tarefas e domínios, incluindo:
* **Visão Computacional**: reconhecimento de objetos, detecção de objetos, segmentação semântica, reconhecimento facial e síntese de imagens.
* **Processamento   de   Linguagem   Natural   (PLN)**:   tradução   automática,   análise   de sentimentos, geração de texto, resumo automático e reconhecimento de fala.
* **Sistemas  de  Recomendação**:  filtragem  colaborativa,  personalizaçãode  conteúdo  e segmentação de usuários.
* **Jogos**: treinamento de agentes para jogar videogames ou jogos de tabuleiro como Go e xadrez.
* **Robótica**:  navegação  autônoma,  controle  de  manipuladores  e  interação  humano-robô.
* **Análise de Séries Temporais**: previsão de demanda, análise de tendências e detecção de anomalias.

O sucesso do Deep Learning pode ser atribuído a vários fatores, incluindo o aumento da disponibilidade de dados, avanços no poder de computação (especialmente GPUs) e melhorias nas técnicas de treinamento e arquiteturas de modelos."

"Existem vários frameworks Python para construir modelos de Deep Learning. Alguns dos principais são:
* **TensorFlow**: Desenvolvido pelo Google Brain, o TensorFlowé um dos frameworks mais populares e amplamente utilizados para Deep Learning. Ele fornece uma API de alto nível (Keras) para construir e treinar modelos de Deep Learning com facilidade e eficiência.
* **Keras**:  Embora  agora  seja  uma  parte  do  TensorFlow,  o  Keras  começou  como  uma biblioteca  de  alto  nível  independente  para  construir  redes  neurais.  Ele  oferece  uma  interface simples e intuitiva, tornando-o uma escolha popular para iniciantes e especialistas.
* **PyTorch**: Desenvolvido pelo Facebook AI Research, o PyTorch é outro framework popular e versátil para construir modelos de Deep Learning. Ele oferece uma abordagem mais "Pythonic" e  dinâmica  para  criação  e  treinamento  de  modelos,  com  excelente  suporte  para  pesquisa  e desenvolvimento.
* **MXNet**:  Desenvolvido  pela  Apache  Software  Foundation,  o  MXNet  é  um  framework flexível e eficiente que suporta uma variedade de linguagens de programação, incluindo Python. Ele oferece escalabilidade e desempenho em várias GPUs e computação distribuída.
* **PaddlePaddle**:  Desenvolvido  pela  Baidu,  o  PaddlePaddle  é  um  framework  de  Deep Learning que oferece uma plataforma fácil de usar e eficiente para desenvolver e treinar modelos de aprendizado profundo.

Esses são apenas alguns dos principais frameworks disponíveis para construir modelos de Deep  Learning  em  Python.  A  escolha  do  framework  apropriado  dependerá  das  necessidades específicas do projeto e das preferências pessoais do desenvolvedor.

O  uso  de  um  framework  é  opcional,  mas  facilita  bastante  a  construção  de  modelos  de Deep Learning. Neste capítulo usaremos o TensorFlow com o Keras."

**Modelo de Execução do TensorFlow**

O aprendizado de máquina pode se tornar complexo rapidamente e os modelos de aprendizado profundo podem se tornar grandes. Para muitos modelos, você precisa de treinamento distribuído para poder iterar dentro de um prazo razoável. E você normalmente deseja que os modelos desenvolvidos sejam implantados em várias plataformas.

Com a versão atual do TensorFlow, você escreve código para construir um grafo computacional e, em seguida, executá-lo. O grafo é uma estrutura de dados que descreve completamente a computação que você deseja executar. Isso tem muitas vantagens:

* É portátil, já que o grafo pode ser executado imediatamente ou salvo para uso posterior, e pode ser executado em várias plataformas: CPUs, GPUs, TPUs, assim como dispositivos móveis e embarcados. Além disso, ele pode ser implementado em produção sem depender de nenhum código que criou o grafo.
* É transformável e otimizável, pois o grafo pode ser transformado para produzir uma versão mais otimizada para uma determinada plataforma. Além disso, otimizações de memória ou computação podem ser realizadas e compensações feitas entre elas. Isso é útil, por exemplo, no suporte à inferência móvel mais rápida após o treinamento em máquinas maiores.
* Suporte para execução distribuída.
* As APIs de alto nível do TensorFlow, em conjunto com os grafos computacionais, permitem um ambiente de desenvolvimento rico e flexível e poderosos recursos de produção no mesmo framework.

```python
# Função para exibir as imagens
def visualiza_imagens(images, labels): # recebe as imagens e os labels
    plt.figure(figsize = (10,10)) # define o tamanho da área de plotagem
    for i in range(25): # 25 imagens
        plt.subplot(5, 5, i+1) # 5 imagens, 5 linhas
        plt.xticks([]) # omitir os ticks (marcações) em x e y
        plt.yticks([])
        plt.grid(False) # sem gride
        plt.imshow(images[i], cmap = plt.cm.binary) # imshow para mostrar as imagens
        # cada imagem no índice 'i' ; cmap - mapa de coloração
        plt.xlabel(nomes_classes[labels[i][0]]) # no label de x, o nome da classe
    plt.show()
```

```python
# Executa a função
visualiza_imagens(imagens_treino, labels_treino)
```

Matriz para cada imagem, e dentro das matrizes temos valores numéricos representando os pixels. Cada pixel representa a coloração naquele ponto específico da imagem.

Passar a matriz por um conjunto de operações matemáticas.
A primeira é de Convolução, uma outra matriz que vai passar por toda a matriz do dado de entrada (imagem), e vai realizando multiplicações, no final terá uma matriz resultante!Pega essa matriz resultante e passa em uma função de ativação (vários tipos, garanti o aprendizado mesmo quando não há linearidade) Relu, saída! Essa saída tem uma determinada dimensão, muito alta ainda, então, vamos fazer uma redução de dimensionalidade com uma camada de Pooling (vários tipos também), reduzindo o resultado. Passa a saída do Pooling por outra camada de Convolução, Relu,... nesse ciclo enquanto eu quiser que o modelo aprenda alguma coisa.
* Vai depender de quantas camadas colocar no modelo.
* Quanto mais camadas colocar nesse meio, está aumentando as camadas intermediárias.
* Quanto mais camadas, mais profundo é o modelo.  

Até então teve o Aprendizado dos Atributos, é onde o modelo está aprendendo, extraindo os recursos. Nesse caso é a Arquitetura Convolucional (um dos melhores resultados para visão computacional). Quando passa pelas camadas intermediárias, no final teremos uma matriz,mas não podemos usar ela para a classificação, precisa adicionar camadas específicas para classificar os dados.
Flatten - achatamento! Transformar o objeto para unidimensional! (matriz para um vetor). Depois, com o vetor, aplicar um modelo trivial de machine learning... no final aplicar a Softmax (outra função matemática que vai entregar as probabilidades)

```python
# Modelo

# Cria o objeto de sequência de camadas
modelo_dsa = models.Sequential()

# camada = operação matemática

# Adiciona o primeiro bloco de convolução e max pooling (camada de entrada)
modelo_dsa.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (32, 32, 3))) # com a Relu
modelo_dsa.add(layers.MaxPooling2D((2, 2))) # camada de Pooling, matriz 2x2, que vai aplicar redução de dimensionalidade na saída da Relu

# add - adicionar a camada
# chamar a camada 'Conv2D', do pacote layers, definir o número de neurônios (32) (camada de entrada)
# e definir o tamanho do kernel (matriz de convolução)(3,3), que vai varendo
# toda a imagem, para cada grupo de pixel ele faz a multiplicação e vai gerando o resultado
# input_shape = (32, 32, 3) -- shape da matriz de entrada
# '32' po '32' pixels, comprimento e largura, com '3' canais de cores
# cada imagem é um grupo de 3 matrizes, cada matriz com 32x32, cada qual com a coloração (RGB)

# Adiciona o segundo bloco de convolução e max pooling (camada intermediária)
modelo_dsa.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add(layers.MaxPooling2D((2, 2)))
# Obs: input_shape = (32, 32, 3) -- só na camada de entrada
# A saída do 2º bloco é ainda menor

# Adiciona o terceiro bloco de convolução e max pooling (camada intermediária)
modelo_dsa.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add(layers.MaxPooling2D((2, 2)))

# pode ter várias camadas intermediárias!
```

```python
# Adicionar camadas de classificação
modelo_dsa.add(layers.Flatten()) # flatten para unidimensionar. achatar
modelo_dsa.add(layers.Dense(64, activation = 'relu')) # camada densa, terminar de organizar o aprendizado do modelo
modelo_dsa.add(layers.Dense(10, activation = 'softmax')) # camada densa, ativação softmax, que vai entregar
# as probabilidades. '10' pq temos 10 classes, probabilidade de 10 classes
```

```python
# Sumário do modelo
modelo_dsa.summary()
# o modelo ainda não está treinado, só a 'casca', a estrutura
```

Machine Learning vai aproximar uma equação matemática que explique os relacionamentos (entre a imagem e a classificação, por exemplo).
À medida que vamos treinando o modelo, ele vai encontrando a aproximação da função. A função tem vários valores numéricos,  que são os parâmetros... coeficientes!

Todo modelo vai ter um número de parâmetros que serão aprendidos no decorrer do treinamento. Mai de 73.000 nesse caso.

Quanto maior o número de parâmetros, mais preciso é o modelo!

```python
# Compilação do modelo
modelo_dsa.compile(optimizer = 'adam',
                   loss = 'sparse_categorical_crossentropy',
                   metrics = ['accuracy'])
# compilar o modelo usando o adam
# adam - algoritmo
# loss - definir uma função de erro (sparse_categorical_crossentropy),
# reduzindo essa função (otimizar a função)
# metrics = ['accuracy'] - métrica da acurácia para analisar o modelo
```

Algoritmo específico para o aprendizado, que aplica a otimização através do backpropagation (década de 80)!

Um deles é o Adam!
